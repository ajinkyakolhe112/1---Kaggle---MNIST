```python
import datasets as huggingface_datasets
import matplotlib.pyplot as plt

dataset_splits = huggingface_datasets.load_dataset("mnist")
train_dataset, test_dataset = dataset_splits['train'], dataset_splits['test']
```

```python
plt.figure(figsize=(2,2))
plt.imshow(train_dataset[0]['image'], cmap="gray")

# figsize=(600*px, 200*px)
# figsize=(15*cm, 5*cm)
```

```python
from torchvision import transforms as transforms_lib
def transforms(example):
    example['tensor_values'] = transforms_lib.ToTensor()(example['image'])
    return example
train_dataset = train_dataset.map(transforms, batched=False)
```

```python
import torch
import torch.nn as nn

class Network:
    def initialize(self):
        pass
    def forward_pass(self):
        pass
    def backward_pass(self):
        pass

# Architecture CNN ()
class Baseline(nn.Module):
    def __init__(self):
        super().__init__()
        self.feature_extractor = nn.Sequential(
            nn.Conv2d(out_channels = 30, in_channels = 1, kernel_size=(3,3)),
            nn.ReLU(),
            nn.Conv2d(out_channels = 40, in_channels = 30, kernel_size=(3,3)),
            nn.ReLU(),
            nn.Conv2d(out_channels = 50, in_channels = 40, kernel_size=(3,3)),
            nn.ReLU(),
        )
    
    def forward(self , examples_batch):
        final_output = self.feature_extractor(examples_batch)
        return final_output
```
